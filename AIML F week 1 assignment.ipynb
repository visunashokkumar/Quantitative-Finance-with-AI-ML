### Importing libraries

```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import yfinance as yf
```

```python
start = "2015-01-28"
end = '2022-01-28'
stk1 = 'RELIANCE.NS'
stk2 = 'HDFCBANK.NS'
stk3 = 'TITAN.NS'

asset1 = yf.download(stk1 ,start,end)
asset2 = yf.download(stk2 ,start,end)
asset3 = yf.download(stk3 ,start,end)
```

```python
asset1['Adj Close'].plot(label = stk1 , figsize = (18,7.5))
plt.title(stk1)
```

```python
asset2['Adj Close'].plot(label = 'S&P 500', figsize = (18,7.5))
plt.title(stk2)
```

```python
asset3['Adj Close'].plot(label = 'Feds 10y interest rate', figsize = (18,7.5))
plt.title(stk3)
```

```python
closePxAsset1 = asset1['Adj Close']
closePxAsset2 = asset2['Adj Close']
closePxAsset3 = asset3['Adj Close']
```

```python
mavgAsset1200 = closePxAsset1.rolling(window=200).mean()
mavgAsset2200 = closePxAsset2.rolling(window=200).mean()
mavgAsset3200 = closePxAsset3.rolling(window=200).mean()
```

```python
mavgAsset150 = closePxAsset1.rolling(window=50).mean()
mavgAsset250 = closePxAsset2.rolling(window=50).mean()
mavgAsset350 = closePxAsset3.rolling(window=50).mean()
```

```python
from pandas import DataFrame
from matplotlib import style
```

```python
mpl.rc('figure', figsize = (18,7.5))

style.use('ggplot')

closePxAsset1.plot(label = stk1)
mavgAsset1200.plot(label = '200d MA')
mavgAsset150.plot(label = '50d MA')

plt.legend()
plt.title( stk1 , fontsize = 16)
plt.xlabel('Date', fontsize = 16)
plt.ylabel('Price', fontsize = 16)
plt.grid(True)
plt.show()
```

```python
mpl.rc('figure', figsize = (18,7.5))

style.use('ggplot')

closePxAsset2.plot(label = stk2)
mavgAsset2200.plot(label = '200d MA')
mavgAsset250.plot(label = '50d MA')

plt.legend()
plt.title( stk2 , fontsize = 16)
plt.xlabel('Date', fontsize = 16)
plt.ylabel('Price', fontsize = 16)
plt.grid(True)
plt.show()
```

```python
mpl.rc('figure', figsize = (18,7.5))

style.use('ggplot')

closePxAsset3.plot(label = stk3)
mavgAsset3200.plot(label = '200d MA')
mavgAsset350.plot(label = '50d MA')

plt.legend()
plt.title(stk3, fontsize = 16)
plt.xlabel('Date', fontsize = 16)
plt.ylabel('Price', fontsize = 16)
plt.grid(True)
plt.show()
```

```python
asset1['Volume'].plot(label = stk1 , figsize = (18,7.5))
asset2['Volume'].plot(label = stk2)
asset3['Volume'].plot(label = stk3)
plt.title('Volumes of assets traded')
plt.legend()
```

```python
asset1MCap = asset1['Open'] * asset1['Volume']
asset2MCap = asset2['Open'] * asset2['Volume']
asset3MCap = asset3['Open'] * asset3['Volume']

asset1MCap.plot(label = stk1, figsize = (18,7.5))
asset2MCap.plot(label = stk2 )
asset3MCap.plot(label = stk3 )
plt.title('Market Cap')
plt.legend()
```

## Scatter matrix

```python
data = pd.concat([asset1['Adj Close'], asset2['Adj Close'], asset3['Adj Close']], axis = 1)
data.columns = [ stk1 , stk2, stk3]
pd.plotting.scatter_matrix(data, figsize = (8, 8), hist_kwds = {'bins' : 250})
```

```python
asset1['returns'] = (asset1['Adj Close']/asset1['Adj Close'].shift(1)) -1
asset2['returns'] = (asset2['Adj Close']/asset2['Adj Close'].shift(1))-1
asset3['returns'] = (asset3['Adj Close']/asset3['Adj Close'].shift(1)) - 1
asset1['returns'].hist(bins = 100, label = 'TCS', alpha = 0.5, figsize = (15,7))
asset2['returns'].hist(bins = 100, label = 'Infosysy', alpha = 0.5)
asset3['returns'].hist(bins = 100, label = 'Wipro', alpha = 0.5)
plt.legend()
```

```python
asset1['returns'].plot(label= stk1 )
asset2['returns'].plot(label= stk2 )
asset3['returns'].plot(label= stk3 )
```
